### PPI

```python
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
with open('data/ai.txt', 'r') as f:
    data = f.readlines()
```

```python
import re
```

```python
join_text = ' '.join(data)
join_text = re.sub(r'[^\w\s]','', join_text)
train_token = word_tokenize(join_text.lower())
# toek# word_tokenize(data.lower())
```

### Term-term matrix

```python
word_count = Counter(train_token)
vocab = word_count.most_common(50)
```

```python
# create a dictionary
word_to_index = {}
for i, (word, count) in enumerate(vocab):
    word_to_index[word] = i
```

```python
tokens_index = []
for token in train_token:
    if token in word_to_index:
        tokens_index.append(word_to_index[token])
    else: 
        tokens_index.append(-1) # no existing in most common word

print(tokens_index)
```
