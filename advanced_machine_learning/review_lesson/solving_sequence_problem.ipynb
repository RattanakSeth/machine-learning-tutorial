# Solving the sequence problem with LSTM in keras

```python
import os

os.environ["KERAS_BACKEND"] = "tensorflow"

from numpy import array
# from keras.pre import one_hot
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
# from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
# from keras.preprocessing.text import Tokenizer
from keras.layers import Input
# from keras.layers.merge import Concatenate
from keras.layers import Bidirectional

import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
```

### Creating the Dataset
In this next step, we will prepare the dataset that we are going to use for this section.

```python
X = list()
Y = list()
X = [x+1 for x in range(20)]
Y = [y * 15 for y in X]

X = np.array(X)
Y = np.array(Y)

print(X)
print(Y)
```

The input to the LSTM layer should be in 3D shape i.e. (samples, time-steps, features). The samples are the number of samples in the input data. We have 20 samples in the input. The time-steps is the number of time-steps per sample. We have 1 time-step. Finally, features correspond to the number of features per time-step. We have one feature per time-step.

```python
X = array(X).reshape(20, 1, 1)
print(X[0:3])
```

### Solution via Simple LSTM
Now we can create our simple LSTM model with one LSTM layer.

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```python
# train model
model.fit(X, Y, epochs=2000, validation_split=0.2, batch_size=5)
```

```python
test_input = array([30])
test_input = test_input.reshape((1, 1, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```
