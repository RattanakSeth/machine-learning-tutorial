### Dimensional of the feature map

```python
import tensorflow as tf;
```

```python
# Define input data (4D tensor), e.g., a batch of 1 image (4x4) with 1 channel
input_data = tf.constant([
    [
        [0,0,1,2],
        [0,0,2,0],
        [0,0,0,2],
        [0,0,1,2],
        [0,0,0,2],
        [0,0,2,0],
        [0,0,1,2]
    ]
], dtype=tf.float32)

print("original input data: ", input_data)

# Reshape to match the expected 4D input `[batch, height, width, channels]`
input_data = tf.reshape(input_data, [1, 7, 4, 1])
print(input_data)
```

```python
# Define a 3x3 filter (kernel) with 1 input channel and 1 output channel
filter_weights = tf.constant([
    [[-1,-1,1],
    [-1, -1, 1],
    [-1, -1, 1]]
], dtype=tf.float32)
print(filter_weights)

# Reshape to match the expected filter shape `[filter_height, filter_width, in_channels, out_channels]`
filter_weights = tf.reshape(filter_weights, [3,3,1,1])
print("filter weight after reshape: ", filter_weights)
```

```python
# Strides: [batch_stride, height_stride, width_stride, channel_stride]
strides = [1,2,1,1] # Setting stride to 2 for height and 1 width

# padding can either 'SAME' or 'VALID'
padding = 'VALID'

# Perform 2D convolution
output = tf.nn.conv2d(input=input_data, filters=filter_weights, strides=strides, padding=padding)
```

```python
print(tf.reshape(output, [3,2]))
```
