```python
# Exercise 7.1 with Scikit learn
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/food_consumption.csv').to_numpy()
x = data[:, 1:]
y = data[:, 0]
x.shape
```

```python
# from sklearn import preprocessing

# scaler = preprocessing.StandardScaler().fit(x)

#feature scaling
def feature_scaling(x):
  mu = np.mean(x, axis=0) #np(n)
  x = (x-mu)/(np.max(x, axis=0)-np.min(x, axis=0))
  return x

# x0 = np.ones((x.shape[0], 1))
# x1 = x.reshape(-1, 1)
# x2 = np.sqrt(x).reshape(-1, 1)
# tmp = np.concatenate((x1, x2), axis=1)
x_scaled = feature_scaling(x)
# x_scaled = np.concatenate((x0, x_scaled), axis=1)
# print(x_scaled.shape)
```

```python
x_scaled.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(x_scaled)
new_x = pca.transform(x_scaled)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
print("new_x", new_x.shape)
```

```python
from matplotlib.pyplot import scatter, annotate, show

scatter(x=new_x[:, 0], y=new_x[:, 1])
for i in range(len(y)):
  annotate(y[i], new_x[i])
show()
```
