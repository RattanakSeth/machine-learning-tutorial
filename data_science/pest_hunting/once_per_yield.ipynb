### Choose to spray once per rice yield
- This data has three main variables, such as parasitoid, predator and week of spray to get different crop yield result.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('../../data/pest_hunting/once_per_yield_natural_enemies.csv')
```

```python
data.head()
```

```python
data = np.array(data)
x, y, z, w = data[:,0], data[:,1], data[:,2], data[:,3]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=w, cmap="viridis", s=80)
plt.colorbar(sc, label="Crop Yield")

ax.set_xlabel("Parasitoid")
ax.set_ylabel("Predator")
ax.set_zlabel("Week of Spray")
plt.show()
```
