### Spray every week
- depend on two params here number of parasitoids and predators

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../data/pest_hunting/every_week_natural_enemies.csv')
```

```python
data = np.array(data)
X, Y, Z = data[:,0], data[:,1], data[:,2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, c=Z, cmap='viridis', s=80)

ax.set_xlabel("Number of Parasitoids")
ax.set_ylabel("Number of Predators")
ax.set_zlabel("Crop Yield")
```

```python
plt.scatter(X, Y, c=Z, s=Z*10, cmap="viridis", alpha=0.7)
plt.colorbar(label="Crop Yield")
```
