### WEEK PEST COME TO THE RICE FIELD
- This will combine parameters such as number of parasitoid, number of predator, spray and week that pest arrived together
- Spray variable will start from 1 to 11, indicating a week that was choosing to spray during a rice cycle. Beside, 0 represents no spray and 13 represent spray every week

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../data/pest_hunting/week_pest_come_natural_enemies.csv')
```

```python
cData = pd.DataFrame({'natural_enemy': data['number_of_parasitoid'] + data['number_of_predator'], 'spray_week': data['sprayWeek'], 'week_pest_come': data['week_pest_come'], 'yield': data['finalYield']})
cData
```

```python
idxMax = np.argmax(data['finalYield'])
idxMax
data.iloc[idxMax]
```

```python
x,y,z, w = cData['natural_enemy'], cData['week_pest_come'], cData['spray_week'], cData['yield']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=w, cmap="viridis", s=80)
plt.colorbar(sc, label="Crop Yield")

ax.set_xlabel("Natural enemy")
ax.set_ylabel("Week Pest Come")
ax.set_zlabel("Week of Spray")
plt.show()
```
